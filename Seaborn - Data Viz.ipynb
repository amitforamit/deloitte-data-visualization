```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
#fig,ax = plt.subplots()
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,label = 'Sine',height=4,aspect =2)
plt.legend(loc='upper left')
```

```python
#fig,ax = plt.subplots()
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,label = 'Sine',height=4,aspect =2,hue='event')
plt.legend(loc='right')
```

```python
#fig,ax = plt.subplots()
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,label = 'Sine',style='region',height=4,aspect =2,hue='event')
plt.legend(loc='right')
```
