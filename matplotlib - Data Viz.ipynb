```python
conditions = [(df['Category'] == 'Furniture') , (df['Category'] == 'Office Supplies'),(df['Category'] == 'Technology')]
c_choices = ['green','black','red']
df['color_map'] = np.select(conditions,c_choices)
```

```python
plt.scatter(sales, profit, s = 10, c = df['color_map'],alpha = 0.4)
```

```python
plt.scatter(sales, profit, s = df['Quantity']*5,alpha = 0.4)
plt.xlabel('sales')
plt.ylabel('profit')
```

```python
cat = df['Category']
```

```python
plt.bar(cat,sales)
plt.legend(cat)
plt.show()
```

# subplot

```python
y = randint(1,50,20)
```

```python
x,y
```

```python
y1= y*x
```

```python
y1
```

```python
plt.plot(x,y)
plt.plot(x,y1)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y1)
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.4,0.3])

ax1.scatter(x,y,color='r')

ax2.plot(x,y1,color='b')
```

```python
labels = ['a','b','c','d']
marks = [15,24,23,20]
```

```python
fig,ax = plt.subplots(1,2,figsize=(20,10))
explode = (0,0.2,0,0.1)
ax[0].pie(marks,labels = labels, autopct = '%1.2f%%',startangle = 90,explode = explode)
ax[1].scatter(x,y,color='r')
plt.show()
```
