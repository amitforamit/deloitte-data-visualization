# Matplotlib - Data Visualization

```python
# importing Libraries

import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
```

```python
x = np.linspace(1,10,20)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color = 'r',marker = 'o', ls = '--')
```

```python
plt.plot(x,y,color = 'r',marker = '*',linestyle = '-.')
```

Labels

```python
plt.plot(x,y,color = 'r',marker = '*',linestyle = '-.')
plt.xlabel('x-axis',color = 'r')
plt.ylabel('y-axis')
plt.xticks(color= 'r')
plt.title('Our first matplotlib chart')
```
